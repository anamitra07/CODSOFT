Data Cleaning

```python
columns_to_drop=['RowNumber',	'CustomerId',	'Surname', 'Geography']
data.drop(columns = columns_to_drop, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender']) # using this to remove categorical values
```

```python
data.head()
```

Modelling

Logistic Regression

```python
from sklearn .model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
X=data.drop(['Exited'], axis=1)
y=data['Exited']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = ohe.fit_transform(X_train[['Gender']])

X_train_numerical = X_train.drop('Gender', axis=1)

X_train_processed = np.hstack((X_train_encoded.toarray(), X_train_numerical.values))

model_log = LogisticRegression()
model_log.fit(X_train_processed, y_train)
```

```python
X_test_encoded = ohe.transform(X_test[['Gender']])

X_test_numerical = X_test.drop('Gender', axis=1)

X_test_processed = np.hstack((X_test_encoded.toarray(), X_test_numerical.values))

y_pred_log = model_log.predict(X_test_processed)
```

```python
accuracy_log = accuracy_score(y_test, y_pred_log)
report_log= classification_report(y_test, y_pred_log)
```

```python
print("Accuracy:",accuracy_log)
print("Classification Report:\n",report_log)
```
